```python
from pyspark.sql import functions as F
import pandas as pd

def data_eda(data):
    stats = []
    for col in data.columns:
        unique_count = data.select(col).distinct().count()
        
        missing_count = data.filter(F.col(col).isNull()).count()
        missing_percentage = round((missing_count / data.count()) * 100, 3)
        
        max_attr_percentage = data.groupBy(col).count().orderBy(F.desc("count")).first()["count"] / data.count() * 100
        
        data_type = dict(data.dtypes)[col]
        
        stats.append((col, unique_count, missing_percentage, round(max_attr_percentage, 3), data_type))
    
    # Converting results to pandas DataFrame
    stats_df = pd.DataFrame(stats, columns=['Column Name', 'Unique Values', 'Missing Value %', 'Max Attribute %', 'Data Type'])
    
    # Sorted by percentage of missing values
    stats_df = stats_df.sort_values('Missing Value %', ascending=False)
    
    return stats_df

```

Handling outliers on the basis of cleaning up missing values

```python
# Initialize a Spark session
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("OutlierHandling").getOrCreate()

# Load the cleaned dataset
data1_clean = spark.read.csv('Demographic_Characteristics_Clean.csv', header=True, inferSchema=True)

# Store the original row count
original_row_count_data1 = data1_clean.count()

# Validate 'Sex' (only values 0 and 1 are valid)
data1_clean = data1_clean.filter(col('Sex').isin([0, 1]))

# Validate 'Age' (values between 1 and 13 are valid)
data1_clean = data1_clean.filter((col('Age') >= 1) & (col('Age') <= 13))

# Validate 'Education' (values between 1 and 6 are valid)
data1_clean = data1_clean.filter((col('Education') >= 1) & (col('Education') <= 6))

# Validate 'Income' (values between 1 and 8 are valid)
data1_clean = data1_clean.filter((col('Income') >= 1) & (col('Income') <= 8))

# Print results
print("Original row count in data1:", original_row_count_data1)
print("Row count in data1 after removing outliers:", data1_clean.count())

# Save the cleaned dataset
data1_clean.write.csv('Demographic_Characteristics_Outlier_Cleaned.csv', header=True)

# Stop the Spark session
spark.stop()
```

Add the variable Comorbidity Index to data3

```python
# Import necessary modules
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when

# Initialize a Spark session
spark = SparkSession.builder.appName("ComorbidityIndex").getOrCreate()

# Load the cleaned data
data3 = spark.read.csv('./Physiological_Health_Indicators_Outlier_Cleaned.csv', header=True, inferSchema=True)

# Check if all necessary columns are present
expected_cols = ['HighBP', 'HighChol', 'Stroke', 'HeartDiseaseorAttack']
missing_cols = [col for col in expected_cols if col not in data3.columns]
if missing_cols:
    raise ValueError(f"One or more required columns are missing from the dataset: {', '.join(missing_cols)}")

# Validate that all values are either 0 or 1
for col_name in expected_cols:
    invalid_rows = data3.filter(~col(col_name).isin(0, 1)).count()
    if invalid_rows > 0:
        raise ValueError(f"Data in column {col_name} contains values other than 0 or 1.")

# Create the Comorbidity Index
data3 = data3.withColumn('ComorbidityIndex', col('HighBP') + col('HighChol') + col('Stroke') + col('HeartDiseaseorAttack'))

# Verify that the Comorbidity Index ranges from 0 to 4
invalid_comorbidity_index = data3.filter(~col('ComorbidityIndex').between(0, 4)).count()
if invalid_comorbidity_index > 0:
    raise ValueError("Comorbidity Index contains values outside the range 0-4.")

# Show the head of the dataset to confirm the new column is added correctly
data3.show(5)

# Save the modified dataset
data3.write.csv('./Physiological_Health_Indicators_With_Comorbidity_Index.csv', header=True)

# Stop the Spark session
spark.stop()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
!pip install imbalanced-learn

# Create result directory if it doesn't exist
os.makedirs('result', exist_ok=True)

def plot_pie_tar(data, col, title):
    plt.figure(figsize=(8, 8))  # Adjust the figure size if necessary
    plt.pie(data[col].value_counts().values,
            labels=data[col].value_counts().index,
            labeldistance=0.8,
            autopct='%1.2f%%',
            pctdistance=1,
            colors=['#00aaff','#aaff00','#ffaa11','#aa99ff'],
            startangle=90,
            )
    plt.title(title)
    plt.savefig(f"result/{title}.jpg", dpi=300)
    plt.show()

# Initialize a Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("SMOTEExample").getOrCreate()

# Load the dataset
data = spark.read.csv('/home/ubuntu/Iteration-4-BDAS/Fully_Merged_Dataset_Cleaned_Inner_Join.csv', header=True, inferSchema=True)

# Convert to Pandas DataFrame for SMOTE processing
data_pd = data.toPandas()

# Plot the pie chart for 'Diabetes_012' column before SMOTE
plot_pie_tar(data_pd, 'Diabetes_012', "Diabetes_012 Before SMOTE")

# Separate features and target
X = data_pd.drop('Diabetes_012', axis=1)
y = data_pd['Diabetes_012']

# Apply SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert back to Pandas DataFrame
resampled_data_pd = pd.DataFrame(X_resampled, columns=X.columns)
resampled_data_pd['Diabetes_012'] = y_resampled

# Plot the pie chart for 'Diabetes_012' column after SMOTE
plot_pie_tar(resampled_data_pd, 'Diabetes_012', "Diabetes_012 After SMOTE")

# Convert back to Spark DataFrame
resampled_data = spark.createDataFrame(resampled_data_pd)

# Save the resampled dataset
resampled_data.write.csv('/home/ubuntu/Iteration-4-BDAS/Fully_Merged_Dataset_SMOTE.csv', header=True, mode='overwrite')

# Stop the Spark session
spark.stop()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pyspark.sql import SparkSession

def analyze_discrete_variables(df, x, target):
    # Convert Spark DataFrame to Pandas DataFrame
    pandas_df = df.select(x, target).toPandas()
    
    # Create a crosstab
    cross_table = pd.crosstab(pandas_df[x], pandas_df[target])
    
    # Plot the crosstab
    cross_table.plot(kind='bar', stacked=True)
    plt.xlabel(x)
    plt.ylabel('Count')
    plt.title(f'{target} by {x}')
    plt.legend(title=target)
    
    # Save the plot
    plt.savefig(f"/home/ubuntu/Iteration-4-BDAS/result/{x}.jpg", dpi=300)
    plt.show()

def analyze_all(df, cols, target):
    for col in cols:
        analyze_discrete_variables(df, col, target)

# Initialize a Spark session
spark = SparkSession.builder.appName("DiscreteVariableAnalysis").getOrCreate()

# Load the merged dataset
merged_data = spark.read.csv('/home/ubuntu/Iteration-4-BDAS/Fully_Merged_Dataset_Cleaned_Inner_Join.csv', header=True, inferSchema=True)

# Define the columns to analyze and the target column
col_lisan = ['Sex','Age','Education','Income','AnyHealthcare' ,'NoDocbcCost','GenHlth','MentHlth','HighBP', 'HighChol', 'Stroke', 'HeartDiseaseorAttack', 'HighChol' ,'PhysHlth','DiffWalk','HighBP','HIghChol','CholCheck','BMI','Smoker', 'Stroke', 'HeartDiseaseorAttack','PhysActivity','Fruits','Veggies','HvyAlcoholConsump','ComorbidityIndex' ]
col_tar = ['Diabetes_012']

# Ensure the result directory exists
import os
os.makedirs('/home/ubuntu/Iteration-4-BDAS/result', exist_ok=True)

# Analyze all specified columns against the target column
analyze_all(merged_data, col_lisan, col_tar[0])

# Stop the Spark session
spark.stop()

```

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

dir_path = '/home/ubuntu/Iteration-4-BDAS/Fully_Merged_Dataset_Cleaned_Inner_Join.csv/'

file_list = [os.path.join(dir_path, f) for f in os.listdir(dir_path) if f.endswith('.csv')]
df = pd.concat([pd.read_csv(file) for file in file_list])


print('Data read successfully.')
print(df.head())


df_new = df.drop('SEQNO', axis=1)

# （70/30）
train, test = train_test_split(df_new, test_size=0.3, random_state=400)


train.reset_index(inplace=True, drop=True)
test.reset_index(inplace=True, drop=True)

print(f"Train set size: {len(train)}")
print(f"Test set size: {len(test)}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

col_lisan = train.columns.drop('Diabetes_012').tolist()
col_tar = 'Diabetes_012'

loc1 = LogisticRegression(max_iter=1000)
loc1.fit(train[col_lisan], train[col_tar])

test_predictions = loc1.predict(test[col_lisan])

conf_matrix = confusion_matrix(test[col_tar], test_predictions)
accuracy = accuracy_score(test[col_tar], test_predictions)
recall = recall_score(test[col_tar], test_predictions, average='weighted')
f1 = f1_score(test[col_tar], test_predictions, average='weighted')


print("Confusion Matrix:\n", conf_matrix)
print("Accuracy:", accuracy)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Logistic Regression Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml import Pipeline
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, f1_score


spark = SparkSession.builder.appName("Random Forest Example").getOrCreate()


data = spark.read.csv('/home/ubuntu/Iteration-4-BDAS/Fully_Merged_Dataset_Cleaned_Inner_Join.csv', header=True, inferSchema=True)

feature_columns = data.columns
feature_columns.remove('Diabetes_012')

vectorAssembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
vectorizedData = vectorAssembler.transform(data)

labelIndexer = StringIndexer(inputCol="Diabetes_012", outputCol="indexedLabel").fit(vectorizedData)
indexedData = labelIndexer.transform(vectorizedData)

(trainingData, testData) = indexedData.randomSplit([0.7, 0.3], seed=400)


rf = RandomForestClassifier(labelCol="indexedLabel", featuresCol="features")


pipeline = Pipeline(stages=[rf])


rfModel = pipeline.fit(trainingData)


predictions = rfModel.transform(testData)


evaluator = MulticlassClassificationEvaluator(labelCol="indexedLabel", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)


print("Accuracy:", accuracy)


preds = predictions.select("prediction", "indexedLabel").toPandas()


conf_matrix = confusion_matrix(preds["indexedLabel"], preds["prediction"])
accuracy = accuracy_score(preds["indexedLabel"], preds["prediction"])
recall = recall_score(preds["indexedLabel"], preds["prediction"], average='weighted')
f1 = f1_score(preds["indexedLabel"], preds["prediction"], average='weighted')


print("Confusion Matrix:\n", conf_matrix)
print("Accuracy:", accuracy)
print("Recall:", recall)
print("F1 Score:", f1)


plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Random Forest Confusion Matrix")
plt.show()

importances = rfModel.stages[-1].featureImportances
importances_df = pd.DataFrame(list(zip(feature_columns, importances)), columns=["Feature", "Importance"])
importances_df = importances_df.sort_values(by="Importance", ascending=False)

plt.figure(figsize=(15, 10))
sns.barplot(x="Importance", y="Feature", data=importances_df)
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Random Forest Feature Importance')
plt.savefig('/home/ubuntu/Iteration-4-BDAS/result/Random_Forest_Feature_Importance.jpg', dpi=300)
plt.show()

spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from sklearn.metrics import confusion_matrix
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


spark = SparkSession.builder.appName("Logistic Regression Example").getOrCreate()


data = spark.read.csv('/home/ubuntu/Iteration-4-BDAS/Fully_Merged_Dataset_Cleaned_Inner_Join.csv', header=True, inferSchema=True)


feature_columns = data.columns
feature_columns.remove('Diabetes_012')


vectorAssembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
vectorizedData = vectorAssembler.transform(data)

labelIndexer = StringIndexer(inputCol="Diabetes_012", outputCol="indexedLabel").fit(vectorizedData)
indexedData = labelIndexer.transform(vectorizedData)


(trainingData, testData) = indexedData.randomSplit([0.7, 0.3], seed=400)

# Logistic Regression
lr2 = LogisticRegression(labelCol="indexedLabel", featuresCol="features")

# Define pipeline
pipeline = Pipeline(stages=[lr2])

# Train the model
lrModel = pipeline.fit(trainingData)

# Make predictions on the test data
predictions = lrModel.transform(testData)

# Evaluate the model using MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(labelCol="indexedLabel", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Accuracy:", accuracy)

evaluator_recall = MulticlassClassificationEvaluator(labelCol="indexedLabel", predictionCol="prediction", metricName="weightedRecall")
evaluator_f1 = MulticlassClassificationEvaluator(labelCol="indexedLabel", predictionCol="prediction", metricName="f1")

recall = evaluator_recall.evaluate(predictions)
f1 = evaluator_f1.evaluate(predictions)

preds = predictions.select("prediction", "indexedLabel").toPandas()
conf_matrix = confusion_matrix(preds["indexedLabel"], preds["prediction"])

print("Confusion Matrix:\n", conf_matrix)
print("Accuracy:", accuracy)
print("Recall:", recall)
print("F1 Score:", f1)

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=[0, 1, 2], yticklabels=[0, 1, 2])
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Logistic Regression Confusion Matrix")
plt.show()

spark.stop()

```
